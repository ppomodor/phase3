# Part2. 함수 프로그램

```python
#변수 넣기
init = int(input('투자금을 입력하세요'))
date_out = input('매수 날짜를 입력하세요')
date_in = input('매도 날짜를 입력하세요')
number = input('종목코드를 입력하세요')
```

```python
#데이터셋
price_start = fdr.DataReader(f'{number}', f'{date_in}', f'{date_in}')['Close']
price_later = fdr.DataReader(f'{number}', f'{date_out}',f'{date_out}')['Close']
close = pd.concat([price_later, price_start])
price = pd.DataFrame(close)
print(price)
```

```python
#결과문 내기
stake = int(init / price.Close[0])
start = price.Close[0] * stake
last = price.Close[1] * stake
sum = last - start
print(f"당신은 {date_out}에 {stake}주를 {start}원에 매수했습니다. 이후 {date_in}에 {last}원에 매도했습니다. 결과적으로 손실/이익은 {sum}입니다")
```
