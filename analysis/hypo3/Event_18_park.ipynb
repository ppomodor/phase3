```python
stock_dict = {
#    'EG':'037370',
    '비트컴퓨터':'032850',
    '대유플러스':'000300',
    '대유에이텍':'002880',
    '서한':'011370'}
stock_dict
```

```python
long_list = []
for item_code in stock_dict.values():
    close = fdr.DataReader(item_code,"2012-01-20" , "2013-01-20")['Close']
    long_list.append(close)
    
df = pd.concat(long_list, axis=1) # 컬럼 기준 병합 
df.columns = stock_dict.keys() # 컬럼은 stock_dict 의 key 값을 사용
#df = df.reset_index()
df.reset_index(inplace=True)
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
df
```

```python
fig = go.Figure()

for stock in df.columns:
  fig.add_trace(
      go.Scatter(
          x = df.index,
          y = df[stock],
          connectgaps=True,
          name = str(stock)
          ) 
      )

fig.update_xaxes(title_text="Date", title_font_size=12)
fig.update_yaxes(title_text="종가", title_font_size=12)
fig.update_layout(title_text=f'<b> 18대 박근혜 이벤트 발생과 종가 추이<b>',
                    autosize=True)
                    #width=1200,
                    #height = 800,
                    #hoverlabel_font_color='#FFFFFF',
                    #plot_bgcolor="#F5F4F4",
                    #title_font_size = 18,
                    #font_color="#3F3F3F")

fig.update_layout(shapes=
                  [dict(type= 'line',
                        yref= 'paper', y0= 0, y1= 1,
                        xref= 'x',x0="2012-04-03", x1="2012-04-03",
                        line=dict(color="MediumPurple",
                                  width=2,
                                  dash="dash")
                        ),
                  dict(type= 'line',
                        yref= 'paper', y0= 0, y1= 1,
                        xref= 'x', x0="2012-01-25", x1="2012-01-25",
                        line=dict(color="MediumPurple",
                                  width=2,
                                  dash="dash")
                       )
                  ])


# add annotation 1
fig.add_annotation(
      x="2012-01-25"
   # , y=20433
    , y= 6950
    , text=f'2012년 1월 25일 <br>금감원 테마주<br>특별조사반 성립'
    , yanchor='bottom'
    , showarrow=True
    , arrowhead=1
    , arrowsize=1
    , arrowwidth=2
    , arrowcolor="#636363"
    , ax= -90
    , ay=-30
    , font=dict(size=16, color="white", family="Courier New, monospace")
    , align="left"
    , bordercolor='#0d47a1'
    , borderwidth=2
    , bgcolor='#0d47a1'
    , opacity= 1
    ,)



# add annotation 2
fig.add_annotation(
      x="2012-04-03"
   # , y=20433
    , y=5700
    , text=f'2012년 4월 3일 <br>전담 조직 확대'
    , yanchor='bottom'
    , showarrow=True
    , arrowhead=1
    , arrowsize=1
    , arrowwidth=2
    , arrowcolor="#636363"
    , ax=-30
    , ay=-40
    , font=dict(size=16, color="white", family="Courier New, monospace")
    , align="left"
    , bordercolor='#0d47a1'
    , borderwidth=2
    , bgcolor='#0d47a1'
    , opacity=1
    ,)

fig.add_vrect(x0="2012-01-23", x1="2012-01-26", 
               fillcolor='#0d47a1', opacity=0.25, line_width=0)
fig.add_vrect(x0="2012-09-10", x1="2012-09-13", 
             fillcolor='crimson', opacity=0.25, line_width=0)
fig.add_vrect(x0="2012-04-02", x1="2012-04-09", 
             
              fillcolor='#0d47a1', opacity=0.25, line_width=0)
fig.show()
```
