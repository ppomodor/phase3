# Part2. 파일 세팅

## 데이터 쌓기

```python
woori = fdr.DataReader('118000', '2016-06-10', '2017-06-10')
barunson = fdr.DataReader('018700',  '2016-06-10', '2017-06-10')
shinil = fdr.DataReader('002700', '2016-06-10', '2017-06-10')
dsrsteel = fdr.DataReader('069730', '2016-06-10', '2017-06-10')
dsr = fdr.DataReader('155660', '2016-06-10', '2017-06-10')
samsung = fdr.DataReader('005930', '2016-06-10', '2017-06-10')
```

```python
# 시작일을 기준으로 수익률 찾기
woori_dpc = (woori['Close']/(woori['Close'][0]) - 1 )*100
barunson_dpc = (barunson['Close']/(barunson['Close'][0]) - 1 )*100
shinil_dpc = (shinil['Close']/(shinil['Close'][0]) - 1 )*100
dsrsteel_dpc = (dsrsteel['Close']/(dsrsteel['Close'][0]) - 1 )*100
dsr_dpc = (dsr['Close']/(dsr['Close'][0]) - 1 )*100
samsung_dpc = (samsung['Close']/samsung['Close'][0] -1) * 100
```

```python
# 데이터 프레임으로 만들기
df = pd.concat([woori_dpc, barunson_dpc, shinil_dpc, dsrsteel_dpc, dsr_dpc, samsung_dpc], axis=1)
df.columns=['우리들휴브레인', '바른손', '신일산업', 'DSR제강', 'DSR', '삼성전자']
df = df.rename_axis('index').reset_index()
df.rename(columns = {'index':'Date'}, inplace=True)
df['삼성전자'].max
```

# Part3. 시각화

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=df['Date'],y=df['우리들휴브레인'],
             mode='lines', name='우리들휴브레인'))
fig.add_trace(go.Scatter(x=df['Date'],y=df['바른손'],
             mode='lines', name='바른손'))
fig.add_trace(go.Scatter(x=df['Date'],y=df['신일산업'],
             mode='lines', name='신일산업'))
fig.add_trace(go.Scatter(x=df['Date'],y=df['DSR제강'],
             mode='lines', name='DSR제강'))
fig.add_trace(go.Scatter(x=df['Date'],y=df['DSR'],
             mode='lines', name='DSR'))
fig.add_trace(go.Scatter(x=df['Date'],y=df['삼성전자'],
             mode='lines', name='삼성전자',
            line=dict(color='black', width=5)))

fig.update_layout(autosize=True,
                 plot_bgcolor="#e5e5e5")

fig.update_xaxes(title_text="날짜", title_font_size=16)
fig.update_yaxes(title_text="수익률(%)", title_font_size=16)

fig.show()



# fig.update_layout(autosize=True,
#                     # width=1200,
#                     # height = 800,
#                     hoverlabel_font_color='#FFFFFF',
#                     plot_bgcolor="#e5e5e5",
#                     title_font_size = 18,
#                     font_color="#3F3F3F")


# fig.update_xaxes(title_text=None, title_font_size=16)
# fig.update_yaxes(title_text="종가 평균", title_font_size=16)

# # fig.add_trace(go.Box(y=df.Average, name='Sample A',
# #                 marker_color = 'indianred'))
# # fig.add_trace(go.Box(y=df.Average, name = 'Sample B',
# #                 marker_color = 'lightseagreen'))


# fig.show()
```
