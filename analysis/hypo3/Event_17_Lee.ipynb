## 1. Stock Info

### 방법1. 변수로 불러와서 쌓기

```python
# Long_term (12months)
samho_l = fdr.DataReader('010960', '2007-01-20','2008-01-20')
ewha_l = fdr.DataReader('001840',  '2007-01-20','2008-01-20')
homecenter_lee_l = fdr.DataReader('060560', '2007-01-20','2008-01-20')
teoksoo_l = fdr.DataReader('026150', '2007-01-20','2008-01-20')
dongsin_l = fdr.DataReader('025950', '2007-01-20','2008-01-20')
mohenz_l = fdr.DataReader('006920', '2007-01-20','2008-01-20')
aprogen_l = fdr.DataReader('007460', '2007-01-20','2008-01-20')
```

### 2. 반복문으로 불러오기

```python
# Long_term_Close
stock_dict = {
    '삼호개발': '010960',
    '이화공영': '001840',
    '홈센타홀딩스': '060560',
    '특수건설': '026150',
    '동신건설': '025950'}

stock_dict

item_list = []
for item_code in stock_dict.values():
    close = fdr.DataReader(item_code, "2007-01-20", "2008-01-20")['Close']
    item_list.append(close)

df_mb_close_l = pd.concat(item_list, axis=1)
df_mb_close_l.columns = stock_dict.keys()
df_mb_close_l.reset_index(inplace=True)

df_mb_close_l['Date'] = pd.to_datetime(df_mb_close_l['Date'])
df_mb_close_l.set_index('Date', inplace=True)
```

```python
# Long_term_Revenue_rate
stock_dict = {
    'samho': '010960',
    'ewha': '001840',
    'homecenter_lee': '060560',
    'teoksoo': '026150',
    'dongsin': '025950',
    'mohenz': '006920',
    # 'aprogen': '007460'
    }
stock_dict

item_list = []
for item_code in stock_dict.values():
    close = fdr.DataReader(item_code, "2006-12-20", "2008-01-20")['Change']
    item_list.append(close)

df_mb_change_l = pd.concat(item_list, axis=1)
df_mb_change_l.columns = stock_dict.keys()
df_mb_change_l.reset_index(inplace=True)

df_mb_change_l['Date'] = pd.to_datetime(df_mb_change_l['Date'])
df_mb_change_l.set_index('Date', inplace=True)
df_mb_change_l = df_mb_change_l*100
```

# Part3. Table Merge 
- Join MB stock price - event table 

```python
#long_term
df_mb_close_l.reset_index()
df_mb_close_l_event = pd.merge(df_mb_close_l, event_17, how='outer', on='Date')
df_mb_close_l_event = df_mb_close_l_event.sort_values(by='Date')
df_mb_close_l_event
```

```python
#long_term_revenue
df_mb_change_l.reset_index()
df_mb_change_l['MB_average'] = (df_mb_change_l['samho'] + df_mb_change_l['ewha'] + df_mb_change_l['homecenter_lee'] + df_mb_change_l['teoksoo'] + df_mb_change_l['dongsin'] + df_mb_change_l['mohenz'])/6
df_mb_change_l_event = pd.merge(df_mb_change_l, event_17, how='outer', on='Date')
df_mb_change_l_event = df_mb_change_l_event.sort_values(by='Date')
df_mb_change_l_event.head(5)
```

# Part5. 이벤트 성격과 주가 상승/하락 관계

## 1. 분포

```python
#positive_table 
df_mb_change_l_event_edit = pd.read_excel('/content/drive/MyDrive/dait5_Final/Phase3/change_event.xlsx')
df_mb_change_l_event_edit.info() 
```

```python
#Event_positive 행 분리 : 'positive'
df_mb_event_positive = df_mb_change_l_event_edit[df_mb_change_l_event_edit['Event_Positive']=='Positive']
df_mb_event_positive['MB_average'].dropna(axis=0)
df_mb_event_positive.describe()
```

```python
#Event_positive 행 분리 : 'Neutral' 
df_mb_event_neutral = df_mb_change_l_event_edit[df_mb_change_l_event_edit['Event_Positive']=='Neutral']
df_mb_event_neutral['MB_average'].dropna()
df_mb_event_neutral.describe()
```

```python
#Event_positive 행 분리 : 'Negative' 
df_mb_event_negative = df_mb_change_l_event_edit[df_mb_change_l_event_edit['Event_Positive']=='Negative']
df_mb_event_negative['MB_average'].dropna()
df_mb_event_negative.describe()
```

```python
#histogram으로 분포 확인
import plotly.figure_factory as ff

# Add histogram data
positive = df_mb_event_positive['MB_average']
neutral = df_mb_event_neutral['MB_average']
negative = df_mb_event_negative['MB_average']

# Group data together
hist_data = [positive.dropna(), neutral.dropna(), negative.dropna()]
colors = ['#e53935', '#ffd180', '#37474f']
group_labels = ['긍정이벤트', '중립이벤트', '부정이벤트']

# Create distplot with custom bin_size
fig_change_hist = ff.create_distplot(hist_data, group_labels, colors=colors)


#여기서부터는 차트 형식 설정입니다. 
fig_change_hist.update_xaxes(title_text="Change_Rates", title_font_size=12)
fig_change_hist.update_yaxes(title_text="분포", title_font_size=12)
fig_change_hist.update_layout(title_text=f'<b> 17대_이명박 이벤트 성격에 따른 수익률 분포<b>',
                    autosize=False,
                    width=1200,
                    height = 600,
                    hoverlabel_font_color='#FFFFFF',
                    plot_bgcolor="#F5F4F4",
                    title_font_size = 18,
                    font_color="#3F3F3F")


fig_change_hist.show()
```
