```python
import FinanceDataReader as fdr
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import plotly.express as px
import pingouin as pg
from statsmodels.formula.api import ols
import matplotlib.pyplot as pit
import statsmodels
from statsmodels.tsa.stattools import coint
from sklearn.preprocessing import MinMaxScaler
import plotly.graph_objects as go
import scipy.stats as ss
import seaborn as sns
from matplotlib import font_manager, rc
import platform

if platform.system() == 'Windows':
    font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
    rc('font', family=font_name)
```

```python
stock_dict ={
    '세명전기':'017510',
    '코마스인':'045390',
    '미주레일':'078940',
    '케이아이씨':'007460'}

stock_dict
```

```python
dy_list = []
for item_code in stock_dict.values():
    close = fdr.DataReader(item_code, "2002-12-20", "2012-12-20")['Close']
    dy_list.append(close)

df = pd.concat(dy_list, axis=1)
df.columns = stock_dict.keys()
df.reset_index(inplace=True)

df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
df
#df_0607= df_0607*100
```

```python
fig = go.Figure()

for stock in df.columns:
  fig.add_trace(
      go.Scatter(
          x = df.index,
          y = df[stock],
          connectgaps=True,
          name = str(stock)
          ) 
      )

fig.update_xaxes(title_text="Date", title_font_size=12)
fig.update_yaxes(title_text="종가", title_font_size=12)
fig.update_layout(title_text=f'<b>17대 정동영 초장기 종가 추이<b>',
                    autosize=True)
                    #width=1200,
                    #height = 800,
                    #hoverlabel_font_color='#FFFFFF',
                    #plot_bgcolor="#F5F4F4",
                    #title_font_size = 18,
                    #font_color="#3F3F3F")
                    
fig.show()
```
