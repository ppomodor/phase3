# 1. 18대 박근혜 초장기 그래프

```python
stock_park = {
    'EG':'037370',
    '비트컴퓨터':'032850',
    '대유플러스':'000300',
    '대유에이텍':'002880',
    '서한':'011370'}
```

```python
gh_list = []
for item_code in stock_park.values():
    close = fdr.DataReader(item_code, "2007-12-19", "2017-12-19")['Close']
    gh_list.append(close)

df = pd.concat(gh_list, axis=1)
df.columns = stock_park.keys()
df.reset_index(inplace=True)

df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)

fig = go.Figure()

for stock in df.columns:
  fig.add_trace(
      go.Scatter(
          x = df.index,
          y = df[stock],
          connectgaps=True,
          name = str(stock)
          ) 
      )

fig.update_xaxes(title_text="날짜", title_font_size=16)
fig.update_yaxes(title_text="종가", title_font_size=16)
fig.update_layout(autosize=True, plot_bgcolor="#E5E5E5",paper_bgcolor = 'white')
                    #width=1200,
                    #height = 800,
                    #hoverlabel_font_color='#FFFFFF',
                    #plot_bgcolor="#F5F4F4",
                    #title_font_size = 18,
                    #font_color="#3F3F3F")
                    
fig.show()
```
